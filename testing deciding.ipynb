```python
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np
import wave
import sys

samplerate, data = wavfile.read('F:/Programs/SDRSharp/normal_conditions_shorten.wav')
sampleTime = 1e+6/samplerate #microseconds per sample i guess..
threshold = ((min(data)+max(data))/2)

print("sample rate:", samplerate)
print(sampleTime, "microsseconds/sample")
```

```python
print("min value:", min(data))
print("max value:", max(data))
print("middle value:", threshold)
print("size:", len(data))
```

```python
lastdata = 0
cache = [-1,-1]
output = []
index = 1
for value in data:

    if value > threshold: 
        if cache != [-1,-1]: #check if buffer is filled and copy it to output and empty buffer
            output.append(cache)
            cache = [-1,-1]
        
        if lastdata != 1:
            cache[0] = 1;
            cache[1] = index*sampleTime
            lastdata = 1
        
    if value < threshold:
        if cache != [-1,-1]: #check if buffer is filled and copy it to output and empty buffer
            output.append(cache)
            cache = [-1,-1]
        
        if lastdata != 0:
            cache[0] = 0;
            cache[1] = index*sampleTime
            lastdata = 0
    index += 1

print(output)
```

```python

#Data is processed. Another value is added between each value so that there are no diagonal lines 
testOutput = []
index = 0
for value in output:
    testOutput.append(value)
    if index != len(output)-1: #not last
        if value[0] == 0:
            testOutput.append([1,value[1]])
        if value[0] == 1:
            testOutput.append([0,value[1]])
    index += 1
    
print(testOutput)



newOutput_y = [0,1]
newOutput_x = [0,0]

index = 0
for value in testOutput: #reorder arrays, super ugly
    newOutput_y.append(value[0])
    newOutput_x.append(value[1])
    index += 1
```

```python
%matplotlib qt 
#open plot in new window, not in notepad

plt.plot(newOutput_x, newOutput_y)
```
